# Primeiro Caderno com Matplotlib

Esse caderno contêm minha primeira interação com Jupyter e Matplotlib.
Ele apresenta um gráfico estilizado de notas de matemática que são
geradas aleatoriamente.

O número de provas (`NUM_EXAMS`) e a nota máxima de uma prova
(`MAX_GRADE`) podem ser modificados em suas respectivas constantes
na seção de [Criação de Dados](#Criação-dos-Dados).

## Criação dos Dados

```python
from random import random
```

```python
def create_students_grades(num_exams, max_grade):
    grades = [round(random() * (max_grade), 2) for x in range(num_exams)]
    return grades
```

```python
NUM_EXAMS = 8
MAX_GRADE = 10
```

```python
data = create_students_grades(NUM_EXAMS, MAX_GRADE)
print(data)
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x_axis = list(range(1, NUM_EXAMS + 1))
y_axis = data
print("X Axis:", x_axis)
print("Y Axis:", y_axis)
```

```python
plt.plot(x_axis, y_axis, color='c', marker='o', markersize=10, linewidth=1, linestyle='dashed')
plt.title("Notas de Matemática", color='r')
plt.xlabel("Provas", color='c')
plt.ylabel("Notas", color='orange')
plt.show()
```
